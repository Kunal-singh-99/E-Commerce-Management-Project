```python
import csv
import os
from datetime import datetime
import getpass
from colorama import Fore, Style
from playsound import playsound


def main_menu():
    while True:
        print("=" * 100)
        print("\t\t\t Welcome to E-Commerce Management System")
        print("=" * 100)
        print("Main Menu:")
        print(" 1.  Employee")
        print(" 2.  Customer")
        print(" 3.  Exit")
        print("=" * 100)
        try:
            choice = int(input("üëâ Please Enter Your Choice: "))
            if choice == 1:
                emp_login()
            elif choice == 2:
                cust_menu()
            elif choice == 3:
                print(Fore.MAGENTA + "Thank You for Coming...!!!" + Style.RESET_ALL)
                break
            else:
                print(Fore.YELLOW + "‚ö† Please Enter a Valid Choice...!!!" + Style.RESET_ALL)
        except:
            print(Fore.YELLOW + "‚ö† Please Enter a Valid Choice...!!!" + Style.RESET_ALL)


def view_emp():
    with open("employees.csv", "r", newline="") as file:
        reader = csv.DictReader(file)
        print("\nAvailable usernames and passwords\n")
        print("-" * 80)
        for row in reader:
            print(f" Username: {row['username']} | Password: {row['password']} | ")
        print("-" * 80)


def add_emp():
    with open("employees.csv", "a", newline="") as file:
        fieldnames = ["username", "password"]
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        if file.tell() == 0:
            writer.writeheader()
        user = input("Enter the Username :")
        Pass = input("Enter the Password :")
        writer.writerow({"username": user, "password": Pass})
    print(Fore.GREEN + "‚úÖ Successfully Added New Employee Login!" + Style.RESET_ALL)


def upd_emp():
    lst = []
    update = False

    with open("employees.csv", "r", newline="") as file:
        reader = csv.DictReader(file)
        User = input("Enter username :")
        for row in reader:
            if row['username'] == User:
                new_pass = input("Please Enter new Password :")
                row['password'] = new_pass
                update = True
            lst.append(row)

    if update:
        with open("employees.csv", "w", newline="") as file2:
            fieldnames = ['username', 'password']
            writer = csv.DictWriter(file2, fieldnames=fieldnames)
            writer.writeheader()
            writer.writerows(lst)
        print(Fore.GREEN + f"‚úÖ {User}'s Password Updated Successfully" + Style.RESET_ALL)
    else:
        print(Fore.YELLOW + f"Username ({User}) Not found ! " + Style.RESET_ALL)


def del_emp():
    lst = []
    deleted = False

    with open("employees.csv", "r", newline="") as file:
        reader = csv.DictReader(file)
        User = input(" Enter Username to Delete : ")
        for row in reader:
            if row['username'] == User:
                deleted = True
                continue
            lst.append(row)

    if deleted:
        with open("employees.csv", "w", newline="") as file2:
            fieldnames = ['username', 'password']
            writer = csv.DictWriter(file2, fieldnames=fieldnames)
            writer.writeheader()
            writer.writerows(lst)
        print(Fore.GREEN + f"‚úÖ {User} deleted successfully!" + Style.RESET_ALL)
    else:
        print(Fore.YELLOW + f"‚ö† Username ({User}) Not found!" + Style.RESET_ALL)


def emp_man():
    while True:
        print("\n" + "=" * 80)
        print("\t\t\t\tManage Employees")
        print("=" * 80)
        print(" 1. View Employees")
        print(" 2. Add Employee")
        print(" 3. Update Employee")
        print(" 4. Delete Employee")
        print(" 5. Exit")

        try:
            choice = int(input("Enter your choice: "))
        except:
            print("Please enter a valid choice :)")
            continue

        if choice == 1:
            view_emp()
        elif choice == 2:
            add_emp()
        elif choice == 3:
            upd_emp()
        elif choice == 4:
            del_emp()
        elif choice == 5:
            print("Exiting Manage Employees menu...")
            break
        else:
            print("Please enter a valid choice :)")


def emp_menu():
    while True:
        print("\n" + "=" * 80)
        print("\t\t\t\tEmployee Menu")
        print("=" * 80)
        print(" 1.  View all Products")
        print(" 2.  Add new Products")
        print(" 3.  Update a Product's detail")
        print(" 4.  Delete a Product")
        print(" 5.  Manage Employees Login")
        print(" 6.  Sale's Data")
        print(" 7.  Exit")
        print("=" * 80)

        try:
            emp_choice = int(input("üëâ Enter your Choice: "))
            if emp_choice == 1:
                view_product()
            elif emp_choice == 2:
                add_product()
            elif emp_choice == 3:
                upd_product()
            elif emp_choice == 4:
                del_product()
            elif emp_choice == 5:
                emp_man()
            elif emp_choice == 6:
                view_sales()
            elif emp_choice == 7:
                
                print(Fore.MAGENTA + "Thank You " + f"{username}" + " for Coming...!!!" + Style.RESET_ALL)
                break
            else:
                print(Fore.YELLOW + "‚ö† Please Enter a Valid Choice" + Style.RESET_ALL)
        except:
            print(Fore.YELLOW + "‚ö† Please Enter a Valid Choice" + Style.RESET_ALL)


def cust_menu():
    customer_name = input("üë§ Please Enter your name: ")
    while True:
        print("\n" + "=" * 80)
        print("\t\t\t\tCustomer Menu")
        print("=" * 80)
        print(" 1.  View Products")
        print(" 2.  Buy Product")
        print(" 3.  View Cart")
        print(" 4.  Delete a Product from Cart")
        print(" 5.  Checkout (Bill)")
        print(" 6.  Past Orders")
        print(" 7.  Exit")
        print("=" * 80)

        try:
            cust_choice = int(input("üëâ Enter your choice: "))
            if cust_choice == 1:
                view_product()
            elif cust_choice == 2:
                buy_product(customer_name)
            elif cust_choice == 3:
                view_cart(customer_name)
            elif cust_choice == 4:
                del_cart(customer_name) 
            elif cust_choice == 5:
                checkout(customer_name)
            elif cust_choice == 6:
                past_orders(customer_name)
            elif cust_choice == 7:
                print(Fore.MAGENTA + "Thank You " + f"{customer_name}" + " for Coming...!!!" + Style.RESET_ALL)
                break
            else:
                print(Fore.YELLOW + "‚ö† Please Enter a Valid Choice" + Style.RESET_ALL)
        except:
            print(Fore.YELLOW + "‚ö† Please Enter a Valid Choice" + Style.RESET_ALL)

def past_orders(customer_name):
    customer_orders = []

    try:
        with open("Sales.csv", "r") as file:
            reader = csv.DictReader(file)
            for row in reader:
                if row['Customer Name'] == customer_name:
                    customer_orders.append(row)

        if not customer_orders:
            print(Fore.YELLOW + "You haven't placed any orders yet." + Style.RESET_ALL)
            return

        print("\n" + "*" * 180)
        print(f"\n\t\t\t\t\t\t\t {customer_name}'s Past Orders\n")
        print("-" * 180)
        for row in customer_orders:
            print(
                f" Time of Purchase: {row['DateTime']} | Customer Name: {row['Customer Name']} | "
                f"Product ID: {row['Product ID']} | Product Name: {row['Product Name']} | "
                f"Quantity: {row['Quantity']} | Price: ‚Çπ{row['Price']} | Sub Total: ‚Çπ{row['Subtotal']}"
            )
        print("-" * 180)

    except FileNotFoundError:
        print(Fore.YELLOW + "‚ö† No sales data found. There are no past orders." + Style.RESET_ALL)

def view_sales():
    try:
        with open("Sales.csv", "r") as file:
            reader = csv.DictReader(file)
            print("\n üìäSale's Data")
            print("-" * 180)
            for row in reader:
                print(
                    f" Time of Purchase: {row['DateTime']} | Customer Name: {row['Customer Name']} | "
                    f"Product ID: {row['Product ID']} | Product Name: {row['Product Name']} | "
                    f"Quantity: {row['Quantity']} | Price: ‚Çπ{row['Price']} | Sub Total: ‚Çπ{row['Subtotal']}"
                )
            print("-" * 180)

    except FileNotFoundError:
        print(Fore.YELLOW + "‚ö† No sales data found." + Style.RESET_ALL)
        



def emp_login():
    global username
    print("\n" + "-" * 80)
    username = input("üë§ Username: ")
    password = getpass.getpass("üîë Password: ")

    with open("employees.csv", "r", newline="") as file:
        reader = csv.DictReader(file)

        for row in reader:
            if row["username"] == username and row["password"] == password:
                print(Fore.GREEN + "‚úÖ Login Successful!" + Style.RESET_ALL)
                emp_menu()
                return
        print(Fore.RED + "‚ùå Invalid username or password" + Style.RESET_ALL)
        print("Wanna Try Again....?")
        log_ch = input("Yes\t  No\n")
        if log_ch.lower() in ["yes", "no"]:
            if log_ch.lower() == "yes":
                emp_login()
            else:
                main_menu()
        else:
            print(Fore.RED + "‚ùå Invalid Character" + Style.RESET_ALL)

def view_product():
    with open("products.csv", "r", newline="") as file:
        reader = csv.DictReader(file)
        print("\nüì¶ Available Products")
        print("-" * 80)
        for row in reader:
            print(f" ID: {row['Product ID']} | Name: {row['Name']} | "
                  f"Category: {row['Category']} | Price: ‚Çπ{row['Price']} | "
                  f"Stock: {row['Stock']}")
        print("-" * 80)


def add_product():
    fieldnames = ["Product ID", "Name", "Category", "Price", "Stock"]
    with open("products.csv", "a", newline="") as file:
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        if file.tell() == 0:
            writer.writeheader()

       
        try:
            n = int(input("‚ûï Enter How Many Products You Want to Add: "))
        except ValueError:
            print(Fore.RED + "Invalid number entered. Returning to menu." + Style.RESET_ALL)
            return

        for x in range(n):
            p1 = input(" Enter Product ID: ")

            exists = False
            with open("products.csv", "r", newline="") as checkfile:
                reader = csv.DictReader(checkfile)
                for row in reader:
                    if row["Product ID"] == p1:
                        print(Fore.YELLOW + "‚ö† Product with that ID already exists!" + Style.RESET_ALL)
                        exists = True
                        break
            if exists:
                continue

            p2 = input(" Enter Product Name: ")
            p3 = input(" Enter Product Category: ")
            
            try:
                p4 = float(input(" Enter Product Price: "))
                p5 = int(input(" Enter Product Quantity: "))
            except ValueError:
                print(Fore.RED + "Invalid price or quantity. Product not added." + Style.RESET_ALL)
                continue 

            writer.writerow({"Product ID": p1, "Name": p2, "Category": p3, "Price": p4, "Stock": p5})
            print(Fore.GREEN + "‚úÖ Successfully Added New Product!" + Style.RESET_ALL)


def upd_product():
    lst = []
    updated = False

    with open("products.csv", "r", newline="") as file:
        reader = csv.DictReader(file)
        prod_id = input("Enter Product ID : ")
        for row in reader:
            if row['Product ID'] == prod_id:
                print("What do you want to update ?")
                upd_choice = input(" Price or Stock\n")
                if upd_choice.lower() == "price":
                    
                    try:
                        new_price = float(input(f"Enter New Price of the {row['Name']} : "))
                        row['Price'] = str(new_price)
                        updated = True
                    except ValueError:
                        print(Fore.RED + "Invalid price entered. No changes made." + Style.RESET_ALL)
                elif upd_choice.lower() == "stock":
                    
                    try:
                        new_stock = int(input(f"Enter New stock of the {row['Name']} : "))
                        row['Stock'] = str(new_stock)
                        updated = True
                    except ValueError:
                        print(Fore.RED + "Invalid stock quantity. No changes made." + Style.RESET_ALL)
            lst.append(row)

    if updated:
        with open("products.csv", "w", newline="") as file2:
            fieldnames = ['Product ID', 'Name', 'Category', 'Price', 'Stock']
            writer = csv.DictWriter(file2, fieldnames=fieldnames)
            writer.writeheader()
            writer.writerows(lst)
        print(Fore.GREEN + "‚úÖ Successfully updated product !" + Style.RESET_ALL)
    else:
        
        print(Fore.YELLOW + f"‚ö† Product with ID '{prod_id}' not found or no changes made." + Style.RESET_ALL)


def del_product():
    lst = []
    deleted = False

    with open("products.csv", "r", newline="") as file:
        reader = csv.DictReader(file)
       
        del_choice = input("Enter the Product ID : ")
        for row in reader:
           
            if row['Product ID'] == del_choice:
                deleted = True
                continue
            else:
                lst.append(row)

    if deleted:
        with open("products.csv", "w", newline="") as file2:
            fieldnames = ['Product ID', 'Name', 'Category', 'Price', 'Stock']
            writer = csv.DictWriter(file2, fieldnames=fieldnames)
            writer.writeheader()
            writer.writerows(lst)
        print(Fore.GREEN + "‚úÖ Successfully Deleted product !" + Style.RESET_ALL)
    else:
        print(Fore.RED + "‚ùå Please Enter Valid Product ID" + Style.RESET_ALL)


def del_cart(customer_name):
    cart=[]
    deleted=False

    with open("cart.csv","r") as file:
        reader = csv.DictReader(file)

        del_choice = int(input("Enter the Product ID : "))
        for row in reader:
            if row['Customer Name'] == customer_name:
                if row['Product ID'] == str(del_choice):
                    deleted = True
                    continue
                else:
                    cart.append(row)
            else:
                print(Fore.YELLOW + "‚ùå Your Cart is Empty" + Style.RESET_ALL)

                
    if deleted:
        with open("cart.csv","w",newline="") as file2:
            fieldnames = ["Cart ID", "Customer Name", "Product ID", "Product Name", "Quantity", "Price", "Total"]
            writer = csv.DictWriter(file2, fieldnames = fieldnames)
            writer.writeheader()
            writer.writerows(cart)
        print(Fore.GREEN + "‚úÖ Successfully Deleted product !" + Style.RESET_ALL)
    else:
        print(Fore.RED + "‚ùå Please Enter Valid Product ID" + Style.RESET_ALL)

            


def checkout(customer_name):
    
    try:
        cart_items = []
        total = 0.0

        with open("cart.csv", "r", newline="") as f:
            reader = csv.DictReader(f)
            for row in reader:
                if row["Customer Name"] == customer_name:
                    cart_items.append(row)

        if not cart_items:
            print(f"\nüõí No items found in {customer_name}'s cart.\n")
            return

        print("\n" + "=" * 80)
        print(f"\t\t\tüßæ Checkout for {customer_name}")
        print("=" * 80)

        for item in cart_items:
            product_name = item["Product Name"]
            price = float(item["Price"])
            qty = int(item["Quantity"])
            subtotal = price * qty
            total += subtotal
            print(f" {product_name} | Price: ‚Çπ{price} | Qty: {qty} | Subtotal: ‚Çπ{subtotal}")

        print("-" * 80)
        print(f" Total Bill = ‚Çπ{total}")
        print("=" * 80)

        confirm = input("\nDo you want to proceed with payment? (yes/no): ").strip().lower()
        if confirm != "yes":
            print("\n‚ùå Checkout cancelled.\n")
            return
        elif confirm == "yes":
            playsound("Sound.mp3")

        sales_fields = ["DateTime", "Customer Name", "Product ID", "Product Name", "Price", "Quantity", "Subtotal"]
        with open("Sales.csv", "a", newline="") as f:
            writer = csv.DictWriter(f, fieldnames=sales_fields)
            if f.tell() == 0:
                writer.writeheader()
            for item in cart_items:
                writer.writerow({
                    "DateTime": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
                    "Customer Name": customer_name,
                    "Product ID": item["Product ID"],
                    "Product Name": item["Product Name"],
                    "Price": item["Price"],
                    "Quantity": item["Quantity"],
                    "Subtotal": float(item["Price"]) * int(item["Quantity"])
                })

        remaining_cart = []
        with open("cart.csv", "r", newline="") as f:
            reader = csv.DictReader(f)
            for row in reader:
                if row["Customer Name"] != customer_name:
                    remaining_cart.append(row)

        with open("cart.csv", "w", newline="") as f:
            cart_fields = ["Cart ID", "Customer Name", "Product ID", "Product Name", "Quantity", "Price", "Total"]
            writer = csv.DictWriter(f, fieldnames=cart_fields)
            writer.writeheader()
            writer.writerows(remaining_cart)

        print(f"\n‚úÖ Payment successful! Thank you for shopping, {customer_name}.\n")

    except FileNotFoundError:
        print(Fore.YELLOW + f"‚ö† {customer_name}'s cart is empty. Nothing to check out." + Style.RESET_ALL)
    


def buy_product(customer_name):
    product_id = input("Enter Product ID to buy: ")
   
    try:
        quantity = int(input("Enter Quantity: "))
        if quantity <= 0:
            print(Fore.YELLOW + "Quantity must be a positive number." + Style.RESET_ALL)
            return
    except ValueError:
        print(Fore.RED + "Invalid quantity entered." + Style.RESET_ALL)
        return

    products = []
    selected = None
    with open("products.csv", "r", newline="") as file:
        reader = csv.DictReader(file)
        for row in reader:
            if row["Product ID"] == product_id:
                selected = row
            products.append(row)

    if not selected:
        print("‚ö† Product not found.")
        return

    if int(selected["Stock"]) < quantity:
        print("‚ö† Not enough stock available.")
        return

    for row in products:
        if row["Product ID"] == product_id:
            row["Stock"] = str(int(row["Stock"]) - quantity)

    with open("products.csv", "w", newline="") as file:
        fieldnames = ["Product ID", "Name", "Category", "Price", "Stock"]
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(products)

    cart_id = product_id + "_" + customer_name
    price_val = float(selected["Price"])
    total_price = price_val * quantity

    with open("cart.csv", "a", newline="") as file:
        fieldnames = ["Cart ID", "Customer Name", "Product ID", "Product Name", "Quantity", "Price", "Total"]
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        if file.tell() == 0:
            writer.writeheader()

        writer.writerow({
            "Cart ID": cart_id,
            "Customer Name": customer_name,
            "Product ID": product_id,
            "Product Name": selected["Name"],
            "Quantity": quantity,
            "Price": selected["Price"],
            "Total": total_price
        })

    print(f"‚úÖ {quantity} x {selected['Name']} added to cart.")


def view_cart(customer_name):
    try:
        with open("cart.csv", "r", newline="") as file:
            reader = csv.DictReader(file)
            cart_items = [row for row in reader if row["Customer Name"] == customer_name]

        if not cart_items:
            print(Fore.YELLOW + "‚ö† Your cart is empty!" + Style.RESET_ALL)
            return

        print("\n" + "=" * 80)
        print(f"\t\t\t\tüõí {customer_name}'s Cart")
        print("=" * 80)

        total_price = 0.0
        for item in cart_items:
            name = item["Product Name"]
            price = float(item["Price"])
            qty = int(item["Quantity"])
            subtotal = qty * price
            total_price += subtotal
            print(f" {name} | Price: ‚Çπ{price} | Qty: {qty} | Subtotal: ‚Çπ{subtotal}")

        print("-" * 80)
        print(f" Total Bill = ‚Çπ{total_price}")
        print("=" * 80)

    except FileNotFoundError:
        print(Fore.YELLOW + "‚ö† Your cart is empty!" + Style.RESET_ALL)



main_menu()
```
