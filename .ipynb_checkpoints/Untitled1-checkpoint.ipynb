```python
import csv

# ===== Add Event =====
def add_event():
    f = open("events.csv", "a", newline="")
    writer = csv.writer(f)
    event_id = input("Enter Event ID: ")
    name = input("Enter Event Name: ")
    date = input("Enter Event Date: ")
    place = input("Enter Event Place: ")
    writer.writerow([event_id, name, date, place])
    f.close()
    print("‚úÖ Event Added Successfully!")

# ===== View Events =====
def view_events():
    try:
        f = open("events.csv", "r")
        reader = csv.reader(f)
        print("\n=== All Events ===")
        for row in reader:
            print(row)
        f.close()
    except FileNotFoundError:
        print("‚ùå No events found yet!")

# ===== Update Event =====
def update_event():
    try:
        f = open("events.csv", "r")
        reader = list(csv.reader(f))
        f.close()
    except FileNotFoundError:
        print("‚ùå No events file found!")
        return

    event_id = input("Enter Event ID to Update: ")

    updated = False
    for row in reader:
        if row[0] == event_id:
            row[1] = input("Enter New Event Name: ")
            row[2] = input("Enter New Event Date: ")
            row[3] = input("Enter New Event Place: ")
            updated = True

    f = open("events.csv", "w", newline="")
    writer = csv.writer(f)
    writer.writerows(reader)
    f.close()

    if updated:
        print("‚úÖ Event Updated Successfully!")
    else:
        print("‚ùå Event Not Found!")

# ===== Delete Event =====
def delete_event():
    try:
        f = open("events.csv", "r")
        reader = list(csv.reader(f))
        f.close()
    except FileNotFoundError:
        print("‚ùå No events file found!")
        return

    event_id = input("Enter Event ID to Delete: ")

    new_list = []
    deleted = False
    for row in reader:
        if row[0] != event_id:
            new_list.append(row)
        else:
            deleted = True

    f = open("events.csv", "w", newline="")
    writer = csv.writer(f)
    writer.writerows(new_list)
    f.close()

    if deleted:
        print("‚úÖ Event Deleted Successfully!")
    else:
        print("‚ùå Event Not Found!")

# ===== Add Review =====
def add_review():
    f = open("reviews.csv", "a", newline="")
    writer = csv.writer(f)
    name = input("Enter Your Name: ")
    review = input("Enter Your Review: ")
    rating = input("Enter Rating (1-5): ")
    writer.writerow([name, review, rating])
    f.close()
    print("‚úÖ Review Added Successfully!")

# ===== View Reviews =====
def view_reviews():
    try:
        f = open("reviews.csv", "r")
        reader = csv.reader(f)
        print("\n=== All Reviews ===")
        for row in reader:
            print(row)
        f.close()
    except FileNotFoundError:
        print("‚ùå No reviews found yet!")

# ===== Main Menu =====
while True:
    print("\n==== Event Management System ====")
    print("1. Add Event")
    print("2. View Events")
    print("3. Update Event")
    print("4. Delete Event")
    print("5. Add Review")
    print("6. View Reviews")
    print("7. Exit")

    ch = input("Enter Choice: ")

    if ch == "1":
        add_event()
    elif ch == "2":
        view_events()
    elif ch == "3":
        update_event()
    elif ch == "4":
        delete_event()
    elif ch == "5":
        add_review()
    elif ch == "6":
        view_reviews()
    elif ch == "7":
        print("üëã Exiting Program...")
        break
    else:
        print("‚ùå Invalid Choice!")
```
