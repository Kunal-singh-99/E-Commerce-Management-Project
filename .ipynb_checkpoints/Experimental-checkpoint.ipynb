```python
import csv
import os
import datetime
import getpass
from colorama import Fore, Style
    
def emp_login():
    username=input(" Username : ")
    password=getpass.getpass(" Password : ")
    with open("employees.csv", "r") as file:
        reader = csv.DictReader(file)

        for row in reader:
            if row["username"] == username and row["password"] == password:
                print(Fore.GREEN + "✅ Login Successful!" + Style.RESET_ALL)

            else:
                print(Fore.RED + "❌ Invalid username or password" + Style.RESET_ALL)

def view_product():
    with open("products.csv","r") as file:
        reader = csv.DictReader(file)
        print("\nAvailable Products : ")
        print("-"*80)
        for row in reader:
            print(f"ID: {row['ID']} | Name: {row['Name']} | "
            f"Category: {row['Category']} | Price: ₹{row['Price']} | "
            f"Stock: {row['Quantity']}")
        print("-"*80)      


def add_product():
    fieldnames=["ID","Name","Category","Price","Quantity"]
    with open("products.csv","a",newline="") as file:
        writer = csv.DictWriter(file,fieldnames=fieldnames)

        n = int(input("Enter How Many Products You Want to Add : "))
        for x in range(n):
            p1 = input(" Enter Product ID : ")
            exists = False
            with open("products.csv","r") as checkfile:
                reader = csv.DictReader(checkfile)
                for row in reader:
                    if row["ID"] == p1:
                        print(Fore.YELLOW + "⚠ Product with that ID already exists...!!" + Style.RESET_ALL)
                        exists = True
                        break

            if exists:
                continue

            p2=input(" Enter Product Name : ")
            p3=input(" Enter Product Category : ")
            p4=input(" Enter Product Price : ")
            p5=input(" Enter Product Quantity : ")

            writer.writerow({"ID":p1,"Name":p2,"Category":p3,"Price":p4,"Quantity":p5})
            print(Fore.GREEN + "✅ Successfully Added New Product!" + Style.RESET_ALL)

def buy_product():
    with open("products.csv","r") as file:
        reader = list(csv.DictReader(file))
    
        print("\nAvailable Products : ")
        print("-"*80)
        for row in reader:
            print(f"ID: {row['ID']} | Name: {row['Name']}")
        print("-"*80)
    
        order = input("Please Enter The Name of the product to Add it in Your Cart : ")
    
        found = False
        for row in reader:
            if row['Name'].lower() == order.lower():
                print(Fore.GREEN + "✅ Successfully Added To Your Cart!" + Style.RESET_ALL)
                found = True
                break
    
        if not found:
            print(Fore.RED + f"❌ {order} isn't Available" + Style.RESET_ALL)

    
print("="*100)
print("\t\t\t\t","Welcome to E-Commerce Management System")
print("="*100)
print("Main Menu:")
print(" 1.  Employee\n 2.  Customer")
    


try :
    choice=int(input("Please Enter Your Choice : "))
    if choice == 1:
        print("-"*100)
        print(" Please Enter your username and Password")
        emp_login()
        print("-"*100)
        print("Main Menu")
        print(" 1.  View all Products\n 2.  Add new Products\n 3.  Update a Product's detail\n 4.  Delete a Product\n 5.  Add an employee\n 6.  Exit")
        emp_choice=int(input("Enter your Choice : "))


        if emp_choice == 1:
            view_product()

        elif emp_choice == 2:
            add_product()


        else:
            print("please Enter a Vaild Choice")

    elif choice == 2:
        print("-"*100)
        print("Welcome to our E-Commerce Platform")
        cust_name=input("Please Enter your name : ")
        print("Main Menu")
        print(" 1.  View Products\n 2.  Buy Product\n 3.  View Cart\n 4.  Delete a Product\n 5.  Checkout(Bill)\n 6.  Exit")

        cust_choice=int(input("enter your choice = "))

        if cust_choice == 1:
            view_product()

        if cust_choice == 2:
            buy_product()

        else:
            print(Fore.YELLOW +"please Enter a Vaild Choice" + Style.RESET_ALL)
        

    else:
        print(Fore.YELLOW +"Please Enter a Valid Choice...!!!" + Style.RESET_ALL)
except:
    print(Fore.YELLOW +"Please Enter a Valid Choice...!!!" + Style.RESET_ALL)
```
